```python
#importing dependencies

import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
#Gives the path to each csv with relevant info
filepath1 = "books10Kclean.csv"
filepath2 ="bestsellerbooksclean.csv"
filepath3 = "quotesclean.csv"

#creates 3 dataframes with each file
tenk_df = pd.read_csv(filepath1)
best_df = pd.read_csv(filepath2)
quotes_df = pd.read_csv(filepath3)

# tenk_df.head()
# best_df.head()
quotes_df.head()
# tenk_df.dtypes
# best_df.dtypes
# quotes_df.dtypes

```

```python
#Joins datasets

#join book10K dataset with bestselletdataset on title (cleans up duplicates and sets title format too)
tenk_df = tenk_df.rename(columns = {"isbn":"primary_isbn10"})
best_df = best_df.drop_duplicates(subset = "primary_isbn10", keep = "first")
tenk_df= tenk_df.drop_duplicates(subset = "primary_isbn10", keep="first")
tenk_df["title"] = tenk_df["title"].str.title()
best_df["title"]=best_df["title"].str.title()
bookconsolidated_df = best_df.merge(tenk_df, on="title", how ="inner", copy = False)
bookconsolidated_df
# tenk_df.dtypes
# best_df.dtypes
# quotes_df.dtypes
# tenk_df = tenk_df.astype({"primary_isbn10":str})
# tenk_df.dtypes
# test = tenk_df["primary_isbn10"]
# type(test)
```

```python
# Joins datasets with the most popular quotes and cleans up the final table. 

#merges the most popular quote of that author (if it exists for that author). 

booksandquotesdf = bookconsolidated_df.merge(quotes_df, on = "author", how ="inner", copy = False)

# renames columns and organizes
booksandquotesdf = booksandquotesdf[["primary_isbn10_x","title","author","publisher","weeks_on_list","original_publication_year","average_rating","text","likes"]]
booksandquotesdf = booksandquotesdf.rename(columns = {"primary_isbn10_x":"primary_isbn10","text":"Most_popular_quote_from_author","likes":"#_of_likes_on_quote_in_goodreads"})
booksandquotesdf = booksandquotesdf.sort_values("#_of_likes_on_quote_in_goodreads")
booksandquotesdf
```
